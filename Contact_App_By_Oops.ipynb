```python
import re
def name_check(name):
  r=re.match(r'[A-Z,a-z]',name)
  if r == name:
    return True
  else:
    return False

def number_check(number):
  r=re.match(r'\d{10}',number)
  if r == number:
    return True
  else:
    return False

def email_check(email):
  r=re.match(r'\w+\.?\w+@\w+\.[a-z]{3}',email)
  if r == email:
    return True
  else:
    return False
    
class Contact:
    def __init__(self, name, number, email):
        self.name = name
        self.number = number
        self.email = email
        
    def update(self, new_name=None, new_number=None, new_email=None):
        if new_name:
            self.name = new_name
        if new_number:
            self.number = new_number
        if new_email:
            self.email = new_email
            
    def __str__(self):
        return f'{self.name}, {self.number}, {self.email}'

class ContactBook:
    def __init__(self):
        self.contacts = []
        
    def add_contact(self):
        name = input("Enter the name of the new contact : ")
        name_check(name)
        number = input('Enter the phone number of the new contact: ')
        number_check(number)
        email = input('Enter the email of the new contact: ')
        email_check(email)
        if name == name_check(name) and number == number_check(name) and email == email_check(email):
          contact = Contact(name, number, email)
          self.contacts.append(contact)
          print('Contact added successfully.')
        
    def open_contact(self):
        name = input("Enter the name of the contact to open or If You Want to Back to the Main Screen just press 'q': ")
        if name.lower() == 'q':
          self.run()
        else:
          results = [c for c in self.contacts if c.name.lower() == name.lower()]
          if results:
              contact = results[0]
              print(contact)
              choice = input("Press 'u' to update or 'd' to delete the contact: ")
              if choice.lower() == 'u':
                  new_name = input("Enter new name (or press enter to keep current name): ")
                  new_number = input("Enter new number (or press enter to keep current number): ")
                  new_email = input("Enter new email (or press enter to keep current email): ")
                  contact.update(new_name, new_number, new_email)
                  print('Contact updated successfully.')
              elif choice.lower() == 'd':
                  self.contacts.remove(contact)
                  print('Contact deleted successfully.')
          else:
              print('No match found!')
            
    def search_contact(self):
        search_term = input("Enter search term or If You Want to Back to the Main Screen just press 'q': ")
        if search_term.lower() == 'q':
          self.run()
        else:
          results = [c for c in self.contacts if search_term.lower() in c.name.lower()]
          if results:
              print('Search results:')
              for result in results:
                  print(result.name)
          else:
              print('No matching contacts found.')
            
    def quit(self):
        print('Goodbye!')
        quit()
        
    def run(self):
        while True:
            print('''\n*******Welcome to The Contact App*******
            1.Add Contact
            2.Open Contact
            3.Search Contact
            4.Quit''')
            choice = input('\nEnter Your Choice : ')
            if choice == '1':
                self.add_contact()
            elif choice == '2':
                self.open_contact()
            elif choice == '3':
                self.search_contact()
            elif choice == '4':
                self.quit()
                break
            else:
                print('Incorrect choice.')

c=ContactBook()
c.run()
```
